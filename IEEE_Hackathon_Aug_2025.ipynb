```python
from paddleocr import PaddleOCR
import os

ocr = PaddleOCR(use_doc_orientation_classify=False,
                use_doc_unwarping=False,
                use_textline_orientation=False, lang='en')
```

```python
result = ocr.predict(os.path.join(os.getcwd(), 'batch2-0500.jpg'))
```

```python
result
```

```python
for res in result:
  res.save_to_img('result')
```

```python
texts = result[0]['rec_texts']
boxes = result[0]['rec_boxes']
scores = result[0]['rec_scores']
```

```python
from paddleocr.tools.infer.utility import draw_ocr
import cv2
from matplotlib import pyplot as plt

image = cv2.imread(os.path.join(os.getcwd(), 'batch2-0500.jpg'))
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.figure((15, 15))
annotated = draw_ocr(image, boxes, texts, scores)
plt.imshow(annotated)
```
